```python
#we divide overall data into dataframe where profit is negative
#concentrate on those areas to improve it..
loss_df=retail_df[retail_df['Profit']<0]
loss_df
```

```python
loss_df.describe() #display summary of loss
```

```python
total_loss=np.negative(loss_df['Profit'].sum())
print("total loss=%2f" %total_loss)
loss_df.groupby(by='Segment').sum()
```

```python
#More discount leads to more loss so offering less discounts more profit is made..
loss_df.groupby(by='Sub-Category').sum()
```

```python
#These are the top 10 cities where more loss is made..
loss_df.groupby(by='City').sum().sort_values('Profit',ascending=True).head(10)
```

```python
#category wise loss
loss_df.sort_values(['Sales'],ascending=True).groupby(by='Category').mean()
```

```python
#we are taking avergae of category wise loss
loss_df.sort_values(['Sales'],ascending=True).groupby(by='Category').mean()
```
