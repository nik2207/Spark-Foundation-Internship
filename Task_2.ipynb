```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

#we are loading iris dataset
iris=datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
#Now We are finding the optimum number of clusters for k-means classification:
x=iris_df.iloc[:,[0,1,2,3]].values
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

#plot theresult on the graph:
plt.plot(range(1,11),wcss)
plt.title('The elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show
```

```python
#Apply k means on the dataset
kmeans=KMeans(n_clusters=3, init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(x)

```

```python
#Now we visualize the cluster for first two columns
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1],s=100,c='red',label='Iris-setosa')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='Iris-veriscolour')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='Iris-virgincia')
plt.legend()
```
